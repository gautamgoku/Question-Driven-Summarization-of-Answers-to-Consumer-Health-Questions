```python
import csv  

header = ['question', 'answer1', 'answer2', 'answer3', 'answer4']

maxi = 0
f = open('asthma_output.csv', 'w', encoding='UTF8')
writer = csv.writer(f)
# write the header
writer.writerow(header)

for index, row in df.iterrows():
    num = 0
    q = row['Question']
    temp_list = []
    temp_list.append(row['Question'])
    q = q.replace('/','-')
    print(q)
    # filename = q + '.txt'
    q = q.replace(' ', '+')
    q = q.replace('?','')
    # file_exists = os.path.exists(filename)
    # if file_exists:
    #     continue
    # with open(filename, 'a') as the_file:
        # the_file.write(row['Question'])
        # the_file.write('_a')
        # the_file.write(row['MedScape'])
        # the_file.write('\n')
    nurl = url + q
    nf = requests.get(nurl, headers = headers)
    soup = BeautifulSoup(nf.content,'lxml')
    answers = soup.find_all('div', {'class' : 'cont-text'})
    l = len(answers)
    i = 0
    for anchor in answers:
        ans = anchor.text
        tans = ans.lower()
        d = row['Disease'].lower()
        if tans.find(d) and len(temp_list) <= 5:
            print(" ".join(anchor.text.split()))
            # the_file.write(" ".join(anchor.text.split()))
            temp_list.append(" ".join(anchor.text.split()))
            # if i < l-1:
            #     the_file.write("_a")
            print('\n')
        elif tans.find('hiv') and len(temp_list) <= 5:
            print(" ".join(anchor.text.split()))
            # the_file.write(" ".join(anchor.text.split()))
            temp_list.append(" ".join(anchor.text.split()))
            # if i < l-1:
            #     the_file.write("_a")
            print('\n')
        i = i+1
        num = num + 1
    if(num > maxi):
        maxi = num
    if(len(temp_list) < 5):
        while(len(temp_list) < 5):
            temp_list.append(" ")
    writer.writerow(temp_list)
```

```python
import pandas as pd
import csv
import unicodedata
```

```python
file = open('/content/asthma_output.csv', encoding= 'unicode_escape')
csvreader = csv.reader(file)
ansList = []
for i, row in enumerate(csvreader):
    # i is the row number
    if i != 0:
      ans = ""
      for j in range(1, 5):
        if(row[j] != ""):
          # print(row[j])
          ans += str(j) + ". "
          ans += row[j]
          ans += "\n"
      ansList.append(ans)

i = 1
ans_list = []
pipeline_summary = []
ans_list.append("BERT Summarized Answer")
pipeline_summary.append("Pipeline Summary")
for k in ansList:
  if k != "":
    bert_sum = ''.join(bert_model(k, min_length=60))
    pipeline_sum = summarization(k)[0]['summary_text']
    # print(str(i) + ". " + bert_sum + "\n")
    ans_list.append(bert_sum)
    pipeline_summary.append(pipeline_sum)
  i = i+1

with open('/content/asthma_output.csv',  encoding= 'unicode_escape') as read_f:
    with open('Asthma_Summary.csv', 'w', newline='') as write_f:
        csv_reader = csv.reader(read_f)
        csv_writer = csv.writer(write_f)
        i = 0
        for row in csv_reader:
          # if(i < 7):
            row.append(ans_list[i])
            row.append(pipeline_summary[i])
            csv_writer.writerow(row)
            i += 1
```
